```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def imshow(img):
    plt.figure(figsize=(5, 5))
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
original_img = cv2.imread("./images/fly.jpg")
imshow(original_img)
```

```python
gray = cv2.cvtColor(original_img, cv2.COLOR_BGR2GRAY)
imshow(gray)
```

```python
_, th = cv2.threshold(gray, 120, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
imshow(th)
```

```python
bit_wise_not_img = cv2.bitwise_not(th)
imshow(bit_wise_not_img)
```

```python
copy_img = original_img.copy()
contours, hierarchy = cv2.findContours(bit_wise_not_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cv2.drawContours(copy_img, contours, -1, (0, 0, 255), 5)
imshow(copy_img)
```

```python
sorted_contours = sorted(contours, key=cv2.contourArea, reverse=True)
for cnt in sorted_contours:
    perimeter = cv2.arcLength(cnt, True)
    approx = cv2.approxPolyDP(cnt, 0.05*perimeter, True)
    if len(approx) == 4:
        break
copy_img = original_img.copy()
cv2.drawContours(copy_img, approx, -1, (255, 0, 0), 10)
imshow(copy_img)
```

```python
x1, y1 = approx[1][0]
x2, y2 = approx[0][0]
x3, y3 = approx[2][0]
x4, y4 = approx[3][0]
```

```python
cv2.circle(copy_img, (x1, y1), 8, (0, 255, 0), 5)
cv2.circle(copy_img, (x2, y2), 8, (0, 0, 0), 5)
cv2.circle(copy_img, (x3, y3), 8, (255, 0, 0), 5)
cv2.circle(copy_img, (x4, y4), 8, (0, 0, 255), 5)
imshow(copy_img)
```

```python
x, y = max(x1, x2, x3, x4), max(y1, y2, y3, y4)
```

```python
inputPt = np.float32(approx)
outputPt = np.float32([
    [x+10, 0],
    [0, 0],
    [0, y+10],
    [x+10, y+10]
])
M = cv2.getPerspectiveTransform(inputPt, outputPt)
dst = cv2.warpPerspective(original_img, M, (x, y))
imshow(dst)
```
